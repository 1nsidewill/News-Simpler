## Filling Template

##### Intro Page Inputs

```python
# headline = str(input())
# text =  str(input())
# media = str(input())

headline = 'BTS 2020'
text = "2020 TIMELINE OF BTS' HOT ISSUES by Gyuha Yi, Jungmin Yeo, Peter Sun Bin"
media = 'https://post-phinf.pstatic.net/MjAyMDA2MDJfMjA4/MDAxNTkxMDQ5NzM2MDUw.7uPV1sIqAZlOIBMGs2NZAr5WGKajF2cwt3mx4Ivi2I4g.QLHsJvUOVGU7Gx1zEPNn4TZIbUKKQnB0LgQksLNbyDog.JPEG/%EB%B0%A9%ED%83%84%EC%86%8C%EB%85%84%EB%8B%A8_%ED%8E%98%EC%8A%A4%ED%83%80_%EA%B0%80%EC%A1%B1%EC%82%AC%EC%A7%84%282%29.jpg?type=w1200'
```

```python
def fill_template (summarized_news, timeline_template):

    year_list = []
    month_list = []
    day_list = []
    headline_list = []
    text_list = []
    media_list = []
    for i in range(len(summarized_news)):
        year_list.append((summarized_news['date'][i].split('-'))[0])
        month_list.append((summarized_news['date'][i].split('-'))[1])
        day_list.append((summarized_news['date'][i].split('-'))[2])

        headline_list.append(summarized_news['title'][i])
        text_list.append(summarized_news['kobert'][i])
        media_list.append(timeline_template['Media'][i+1])

    intro_row = pd.Series(data={'Headline':str(headline), 'Text':str(text), 'Media':str(media), 'Type':'title'})
    new_df = pd.DataFrame(data={'Year': year_list, 'Month':month_list, 'Day':day_list, 'Headline':headline_list, 'Text':text_list, 'Media':media_list})
    template = clean_template.append(intro_row, ignore_index=True)
    template = template.append(new_df)
    
    return template
```

```python
check_df = fill_template (summarized_news, timeline_template)
check_df.head(3)
```
